```
%sql
with cte_teams as (
  select
    home_team as teams
  from
    ipl_results
  union
  select
    away_team as teams
  from
    ipl_results
),
cte as (
  select
    dates,
    t.teams as teams,
    result,
    row_number() over(
      partition by t.teams
      order by
        dates
    ) as id
  from
    ipl_results r
    join cte_teams t on r.home_team = t.teams
    or r.away_team = t.teams
),
cte_diff as (
  select
    dates,
    teams,
    result,
    id,
    id - row_number() over(
      partition by teams
      order by
        id
    ) as diff
  from
    cte
  where
    result = teams
),
cte_final as (
  select
    *,
    count(1) over(
      partition by teams,
      diff
      order by
        id range between unbounded preceding
        and unbounded following
    ) as streak
  from
    cte_diff
)
select
  t.teams,
  coalesce(max(streak), 0) as max_winning_streak
from
  cte_teams t
  left join cte_final f on f.teams = t.teams
group by
  t.teams
order by
  max_winning_streak desc;
```

```
from pyspark.sql.functions import *
from pyspark.sql import Window

# Assuming you have already loaded your IPL results into a DataFrame named 'ipl_results'

# Create a DataFrame of all unique teams
home_teams = df.select(col("home_team").alias("teams"))
away_teams = df.select(col("away_team").alias("teams"))
all_teams = home_teams.union(away_teams).distinct()

# Define window specification for row numbering within each team
from pyspark.sql.window import Window
team_window = Window.partitionBy("teams").orderBy("dates")

# Add row number within each team to identify streaks
cte = df.withColumn("result", when(col("result") == col("home_team"), col("home_team")).otherwise(None)) \
                 .select("dates", "home_team", "away_team", "result") \
                 .join(all_teams, (col("home_team") == col("teams")) | (col("away_team") == col("teams")), "inner") \
                 .withColumn("id", row_number().over(team_window))

# Calculate the streaks
cte_diff = cte.withColumn("diff", col("id") - row_number().over(team_window.partitionBy("teams").orderBy("id"))) \
              .filter(col("result") == col("teams")) \
              .select("dates", "teams", "result", "id", "diff")

cte_final = cte_diff.withColumn("streak", count("*").over(Window.partitionBy("teams", "diff").orderBy("id").rowsBetween(Window.unboundedPreceding, Window.unboundedFollowing)))

# Aggregate to find maximum winning streak per team
max_streaks = all_teams.join(cte_final, "teams", "left_outer") \
                       .groupBy("teams") \
                       .agg(max("streak").alias("max_winning_streak")) \
                       .orderBy(col("max_winning_streak").desc())

max_streaks.display()
```
